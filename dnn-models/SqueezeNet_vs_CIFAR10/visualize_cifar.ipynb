```python
'''
Visualizing CIFAR-10 dataset for better explanation, motivation, etc. 

'''

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# shortcut to modules provided in tf
datasets = tf.contrib.keras.datasets
```

```python
# Create an indexable list for CIFAR-10 classes

cifar_classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# Load the CIFAR-10 dataset through tf
# Note : could take a few minutes depending on machine...

(x_train, y_train), (x_test, y_test)= datasets.cifar10.load_data()

# Create a function to plot/visualize the images.

def plot_25images(images, labels):
    
    # figure size (5x5 = 25)
    fig = plt.figure(figsize=(10,10))
    
    # plot image grid
    for x in range(5):
        for y in range(5):
            ax = fig.add_subplot(5, 5, 5*y+x+1)
            plt.imshow(images[5*y+x])
            plt.xticks(np.array([]))
            plt.yticks(np.array([]))
            labels = labels.flatten() # to 1-D array for indexing
            plt.xlabel(cifar_classes[labels[5*y+x]])
    plt.show()

# Visualize CIFAR-10

print('Training images : ' + str(x_train.shape[0]) + ' images.')
plot_25images(x_train[:25], y_train[:25])

print(20 * '-')

print('Testing images : ' + str(x_test.shape[0]) + ' images.')
plot_25images(x_test[:25], y_test[:25])
```
